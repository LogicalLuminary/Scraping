```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

#YEARWISE ENGINEERING NIRF SCRAPED 2021-2025

```python
import re
otpt ={}

for ar in range(2021,2026):
  webpage = requests.get(f'https://www.nirfindia.org/Rankings/{ar}/EngineeringRanking.html').text
  soup=BeautifulSoup(webpage,'lxml')
  list_str=[]
  tb=soup.find_all('tbody')
  if(len(tb)>0): tb=tb[0]
  else : continue
  for i in tb:
    list_str.append(i.text)
  final=[]
  for s in list_str:
    s=s[11:]
    text = s
    name_match = re.match(r'^(.*?)More DetailsClose', text)
    name = name_match.group(1) if name_match else ""
    score_match = re.findall(r'(\d+\.\d{2})\d*$', text)
    score = score_match[0] if score_match else ""
    # 3️⃣ Extract location (letters before the score)
    location_match = re.search(r'([A-Za-z\s]+)' + score, text)
    location = location_match.group(1).strip() if location_match else ""
    fixed_location = re.sub(r'(?<!\s)([A-Z])', r' \1', location)
    fixed_location= re.sub(r'\s+', ' ', fixed_location).strip()
    # 4️⃣ Extract rank (leftover number after score)
    rank_match = re.search(score + r'(\d+)', text)
    rank = rank_match.group(1) if rank_match else ""
    final.append([name,score,fixed_location,rank])
  otpt[ar]=final

```

```python
import csv

with open("nirf_data_all_years.csv", "w", newline='', encoding="utf-8") as f:
    writer = csv.writer(f)
    # Write header
    writer.writerow(["Year", "Name", "Score", "Location", "Rank"])

    for year, rows in otpt.items():
        for row in rows:
            writer.writerow([year] + row)
```
